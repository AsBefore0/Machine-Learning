```python
'''
min f(x1, x2, x3) = x1^2 + x2^2 + x3^2
s.t.
    x1*x2 >= 1
    x1*x2 <= 5
    x2 + x3 = 1
    0 <= x1, x2, x3 <= 5
'''
```

```python
# 1. 定义目标函数
def obj_func(p):
    x1, x2, x3 = p
    return x1 ** 2 + x2 ** 2 + x3 ** 2


# 2. 等式约束条件 
constraint_eq = [
    lambda x: 1 - x[1] - x[2]
]

# 不等式约束条件 g(x)<=0
constraint_ueq = [
    lambda x: 1 - x[0] * x[1],
    lambda x: x[0] * x[1] - 5
]
```

```python
from sko.DE import DE

# func: 目标函数
# n_dim: 变量个数
# size_pop: 种群规模
# max_iter: 最大迭代次数
# prob_mut: 变异概率 默认0.3
# F: 变异系数 默认0.5
# lb: 变量下界
# ub: 变量上界
# constraint_eq: 等式约束
# constraint_ueq: 不等式约束
de = DE(func=obj_func, n_dim=3, size_pop=50, max_iter=800, lb=[0, 0, 0], ub=[5, 5, 5],
        constraint_eq=constraint_eq, constraint_ueq=constraint_ueq)
```

```python
best_x, best_y = de.run()
print('best_x:', best_x, '\n', 'best_y:', best_y)
```

```python
print(de.generation_best_Y)   # 每一代的最优函数值
print(de.generation_best_X)  #  每一代的最优函数值对应的输入值
print(de.all_history_Y)  #  每一代每个个体的函数值
```
